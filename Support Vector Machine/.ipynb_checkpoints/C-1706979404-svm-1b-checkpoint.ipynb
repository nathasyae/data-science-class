```python
import numpy as np 
import matplotlib.pyplot as plt 

data_colors = [(1, 0, 0), (0, 0, 1)]

def read_points_file(filename):
  pts = []
  with open(filename, "r") as f:
    for pt in f:
      pt = pt.strip("\n").split()
      pts.append([float(pt[0]), float(pt[1])])
  return pts

def read_data(class_0_file, class_1_file):
  pts_0 = read_points_file(class_0_file)
  pts_1 = read_points_file(class_1_file)

  x = pts_0 + pts_1
  labels = [0] * len(pts_0) + [1] * len(pts_1)
  x = np.array(x)
  return (x, labels)


def plot_data(X_train, y_train, X_test, y_test):
  
  X = np.concatenate((X_train, X_test))
  y = np.concatenate((y_train, y_test))

  colors = get_colors(y)
  colors_train = get_colors(y_train)
  colors_test = get_colors(y_test)


  plt.figure(figsize=(12, 4), dpi=150)

  # Plot all data plot
  plt.subplot(131)
  plt.axis('equal')
  plt.scatter(X[:, 0], X[:, 1], c = colors, s = 10, edgecolors=colors)
  plt.title("Data (100%)")


  # training data plot
  plt.subplot(132)
  plt.axis('equal')
  #plt.axis('off')
  plt.scatter(X_train[:, 0], X_train[:, 1], 
              c = colors_train, s = 10, edgecolors=colors_train)
  plt.title("Training Data (80%)")

  # testing data plot
  plt.subplot(133)
  plt.axis('equal')
  #plt.axis('off')
  plt.scatter(X_test[:, 0], X_test[:, 1], 
              c = colors_test, s = 10, edgecolors=colors_test)
  plt.title("Test Data (20%)")
  plt.tight_layout()
  
  plt.show()

def get_colors(y):
  return [data_colors[item] for item in y]

def plot_decision_function(X_train, y_train, X_test, y_test, clf):
  plt.figure(figsize=(8, 4), dpi=150)
  plt.subplot(121)
  plt.title("Training data")
  plot_decision_function_helper(X_train, y_train, clf)
  plt.subplot(122)
  plt.title("Test data")
  plot_decision_function_helper(X_test, y_test, clf, True)
  plt.show()


def plot_decision_function_helper(X, y, clf, show_only_decision_function = False):
  
  colors = get_colors(y)
  plt.axis('equal')
  plt.tight_layout()
  #plt.axis('off')

  plt.scatter(X[:, 0], X[:, 1], c = colors, s = 10, edgecolors=colors)
  ax = plt.gca()
  xlim = ax.get_xlim()
  ylim = ax.get_ylim()

  # Create grid to evaluate model
  xx = np.linspace(xlim[0], xlim[1], 30)
  yy = np.linspace(ylim[0], ylim[1], 30)
  YY, XX = np.meshgrid(yy, xx)
  xy = np.vstack([XX.ravel(), YY.ravel()]).T
  Z = clf.decision_function(xy).reshape(XX.shape)

  
  if  show_only_decision_function:
    # Plot decision boundary
    ax.contour(XX, YY, Z, colors='k', levels=[0], alpha=0.5,
             linestyles=['-'])
  else :
    # Plot decision boundary and margins
    ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
             linestyles=['--', '-', '--'])
    # Plot support vectors
    ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s = 10,
             linewidth=1, facecolors='k', c = 'k', label='Support Vectors')
 
    plt.legend(fontsize='small')
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score,confusion_matrix
```

```python
def test_report(y_test, model):
    # accuracy: (tp + tn) / (p + n)
    # accuracy = accuracy_score(y_test, model)
    # print('Accuracy: %f' % (accuracy))
    
    # precision tp / (tp + fp)
    precision = precision_score(y_test, model)
    print('Precision: %f' % (precision))
    
    # recall: tp / (tp + fn)
    recall = recall_score(y_test, model)
    print('Recall: %f' % (recall))

    # confusion matrix
    matrix = confusion_matrix(y_test, model)
    print(matrix)
```

Import our dataset files, join them together as one array.

```python
x, labels = read_data("dataset/point_class_0_noise.txt", "dataset/point_class_1_noise.txt")
```

Split the data for train and test with 80-20 ratio. 

```python
# Bagi data menjadi train dan test data dengan rasio 80-20
X_train, X_test, y_train, y_test = train_test_split(x, labels, test_size = 0.2, random_state=0)
```

## 3. Experiment

### a) Training

From the EDA part above, we can conclude that the data has a distribution pattern. We can divide the data to two region, left upper triangle for the red and right lower triangle for the test. Even though there are some noise data, the dataset still can be separated by a linear line. So we'll try to build a model with Linear kernel and adding a few parameters.

```python
print('Training Linear SVM')
# Create a linear SVM classifier 
clf1 = svm.SVC(kernel='linear', C=0.1)
clf2 = svm.SVC(kernel='linear', C=1)
clf3 = svm.SVC(kernel='linear', C=10)
clf4 = svm.SVC(kernel='linear', C=100)
clf5 = svm.SVC(kernel='linear', C=1000)
```

```python
clf6 = svm.SVC(kernel='linear', C=0.01)
clf7 = svm.SVC(kernel='linear', C=0.001)
clf8 = svm.SVC(kernel='linear', C=0.0001)
```

```python
# Train classifier 
clf1.fit(X_train, y_train)
clf2.fit(X_train, y_train)
clf3.fit(X_train, y_train)
clf4.fit(X_train, y_train)
clf5.fit(X_train, y_train)
```

We add a C parameter which is a penalty. The bigger the number, the harder it will try not to missclassify the data (on training). As we can see as the C gets bigger, the margin gets smaller. On the test data, we can see that we get a same number of missclassified data even though we already adjust the C.

Since all the model not really give a significant different in test data, if i have to choose, ill go with the clf2, when the margin not too small and not too big.

### b) Test

Now we are going to test the dataset to see the model performance.

```python
# Make predictions on unseen test data

print('Model 1, c=0.1\n')
clf_predictions_1 = clf1.predict(X_test)
print("Accuracy: {}%".format(clf1.score(X_test, y_test) * 100 ))
test_report(y_test, clf_predictions_1)

print('Model 2, c=1\n')
clf_predictions_2 = clf2.predict(X_test)
print("Accuracy: {}%".format(clf2.score(X_test, y_test) * 100 ))
test_report(y_test, clf_predictions_2)

print('Model 3, c=10\n')
clf_predictions_3 = clf3.predict(X_test)
print("Accuracy: {}%".format(clf3.score(X_test, y_test) * 100 ))
test_report(y_test, clf_predictions_3)

print('Model 4, c=100\n')
clf_predictions_4 = clf4.predict(X_test)
print("Accuracy: {}%".format(clf4.score(X_test, y_test) * 100 ))
test_report(y_test, clf_predictions_4)

print('Model 5, c=1000\n')
clf_predictions_5 = clf5.predict(X_test)
print("Accuracy: {}%".format(clf5.score(X_test, y_test) * 100 ))
test_report(y_test, clf_predictions_5)
```

```python
print('Model 6, c=0.01\n')
clf_predictions_6 = clf6.predict(X_test)
print("Accuracy: {}%".format(clf6.score(X_test, y_test) * 100 ))
test_report(y_test, clf_predictions_6)

print('Model 7, c=0.001\n')
clf_predictions_7 = clf7.predict(X_test)
print("Accuracy: {}%".format(clf7.score(X_test, y_test) * 100 ))
test_report(y_test, clf_predictions_7)

print('Model 8, c=0.0001\n')
clf_predictions_8 = clf8.predict(X_test)
print("Accuracy: {}%".format(clf8.score(X_test, y_test) * 100 ))
test_report(y_test, clf_predictions_8)
```

# Multiclass Classification: SVC

## 1. Preparation

First, import the libraries needed.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score,confusion_matrix
```

Import the Iris dataset.

```python
iris_dataset = datasets.load_iris()
```

Split the data for train and test with 80-20 ratio. 

```python
X = iris_dataset['data']
y = iris_dataset['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

## 2. Exploratory Data Analysis (EDA) & Preprocessing

Display information about the data.

```python
print('Training data has {} data'.format(X_train.shape[0]))
print('Test data size is {} data'.format(X_test.shape[0]))
```

```python
print('Null data in training data: {}'.format(np.isnan(X_train).sum()))
print('Null data in test data: {}'.format(np.isnan(X_test).sum()))
```

Now, scale the feature to bring the variable in a single scale.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

Then, plot the data to get visualization.

```python
data_colors = [(1, 0, 0), (0, 0, 1), (0, 1 , 0)]

def get_colors(y):
  return [data_colors[item] for item in y]

def plot_data(X_train, y_train, X_test, y_test):
  
  X = np.concatenate((X_train, X_test))
  y = np.concatenate((y_train, y_test))

  colors = get_colors(y)
  colors_train = get_colors(y_train)
  colors_test = get_colors(y_test)


  plt.figure(figsize=(12, 4), dpi=150)

  # Plot all data plot
  plt.subplot(131)
  plt.axis('equal')
  plt.scatter(X[:, 0], X[:, 1], c = colors, s = 10, edgecolors=colors)
  plt.title("Data (100%)")


  # training data plot
  plt.subplot(132)
  plt.axis('equal')
  #plt.axis('off')
  plt.scatter(X_train[:, 0], X_train[:, 1], 
              c = colors_train, s = 10, edgecolors=colors_train)
  plt.title("Training Data (80%)")

  # testing data plot
  plt.subplot(133)
  plt.axis('equal')
  #plt.axis('off')
  plt.scatter(X_test[:, 0], X_test[:, 1], 
              c = colors_test, s = 10, edgecolors=colors_test)
  plt.title("Test Data (20%)")
  plt.tight_layout()
  
  plt.show()
```

```python
print("Plotting Data")
plot_data(X_train, y_train, X_test, y_test)
```

## 3. Experiment

## Default SVC Model

```python
print('Training Linear SVM')

from sklearn.svm import SVC

# Create a linear SVM classifier
svc_lin = SVC(kernel='linear', random_state = 0)

# Train classifier 
svc_lin.fit(X_train, y_train)
```

```python
print('Training Polynomial SVM')

svc_pol = SVC(kernel='poly')
svc_pol.fit(X_train, y_train)
```

```python
print('Training RBF SVM')

svc_rbf = SVC(kernel='rbf')
svc_rbf.fit(X_train, y_train)
```

Now we are going to test the dataset to see the model performance.

```python
y_pred_lin = svc_lin.predict(X_test)
y_pred_pol = svc_pol.predict(X_test)
y_pred_rbf = svc_rbf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

def report(y_test, y_pred, model_type, model):
    cm = confusion_matrix(y_test, y_pred)
    print(cm)
    print('Accuracy of {} SVC classification is: {:.2f}\n'.format(model_type, model.score(X_test,y_test)))
```

```python
report(y_test, y_pred_lin, 'linear', svc_lin)
report(y_test, y_pred_pol, 'polynomial', svc_pol)
report(y_test, y_pred_lin, 'rbf', svc_rbf)
```

## With Grid Search (C and gamma)

```python
param_grid = {'C': [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001, 0.00001, 10]}
```

```python
from sklearn.model_selection import GridSearchCV

clf_grid = GridSearchCV(SVC(), param_grid, verbose=1)
clf_grid.fit(X_train, y_train)

# clf = clf_grid.best_estimator_()
print('Best Parameters:\n', clf_grid.best_params_)
print('Best Estimators:\n', clf_grid.best_estimator_)
```

```python
clf_predictions = clf_grid.predict(X_test)
report(y_test, clf_predictions, 'Best GridSearch', clf_grid)
```

## Using PCA & Gridsearch (C and gamma)

```python
iris = datasets.load_iris()
```

```python
X = iris.data[:, :]
y = iris.target
```

```python
print('Number of data points :: {}'.format(X.shape[0]))
print('Number of features :: {}'.format(X.shape[1]))
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(1, figsize = (16, 9))
ax = Axes3D(fig, elev = -150, azim = 110)
X_reduced = PCA(n_components = 3).fit_transform(X_scaled)

ax.scatter(X_reduced[:,0], X_reduced[:,1], X_reduced[:, 2], c = y,
          cmap = plt.cm.Set1, edgecolor = 'k', s = 40)

ax.set_title('First three PCA directions')
ax.set_xlabel('1st eigenvecor')
ax.w_xaxis.set_ticklabels([])
ax.set_xlabel('2nd eigenvecor')
ax.w_xaxis.set_ticklabels([])
ax.set_xlabel('3rd eigenvecor')
ax.w_xaxis.set_ticklabels([])

plt.show()
print('The number of features in the new subspace is ', X_reduced.shape[1])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_reduced, y, test_size = 0.2, random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001, 0.00001, 10]}

clf_grid = GridSearchCV(SVC(), param_grid, verbose=1)
clf_grid.fit(X_train, y_train)

# clf = clf_grid.best_estimator_()
print('Best Parameters:\n', clf_grid.best_params_)
print('Best Estimators:\n', clf_grid.best_estimator_)
```

```python
print('Best estimator accuracy on test set {:.2f}'.format(clf_grid.best_estimator_.score(X_test, y_test)))
```
